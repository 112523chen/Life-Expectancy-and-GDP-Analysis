```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns

raw_data = pd.read_csv("all_data.csv")
print(raw_data.columns)
raw_data.columns = ['Country', 'Year', 'Life Expectancy', 'GDP']
print(raw_data.head())
```

```python
print(raw_data.Country.unique())
print(raw_data.Year.unique())
print(np.mean((raw_data.GDP.unique())))
raw_data['Country'] = raw_data['Country'].replace(['United States of America'],'USA')
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="GDP", y="Country", data=raw_data)
plt.xlabel("GDP in Trillions of U.S. Dollars")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Life Expectancy", y="Country", data=raw_data)
plt.xlabel("Life expectancy at birth (years)")
```

```python
plt.figure(figsize=(8,6))
sns.distplot(x=raw_data.GDP, rug = True, kde=False)
plt.xlabel("GDP in Trillions of U.S. Dollars")
```

```python
plt.figure(figsize=(8,6))
sns.distplot(x=raw_data["Life Expectancy"], rug = True, kde=False)
plt.xlabel("Life Expectancy in Years")
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=raw_data.Year, y=raw_data.GDP, hue=raw_data.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("GDP in Trillions of U.S. Dollars")
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=raw_data.Year, y=raw_data["Life Expectancy"], hue=raw_data.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("Life expectancy at birth (years)");
```

```python
graph = sns.FacetGrid(raw_data, col="Country", col_wrap=3,
                      hue = "Country", sharey = False, sharex = False)
graph = (graph.map(sns.scatterplot,"Life Expectancy", "GDP")
         .add_legend()
         .set_axis_labels("Life expectancy at birth (years)", "GDP in Trillions of U.S. Dollars"));
```
